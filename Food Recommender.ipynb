```python
#importing all the necessary libraries
import pandas as pd
```

```python
Food_ratings = pd.read_csv("ratings.csv")
Food_names = pd.read_csv("Foods.csv")
Food_table = pd.merge(Food_ratings, Food_names).drop(['ingredients', 'timestamp'], axis = 1)
Food_table
```

```python
Food_ratings = Food_table.pivot_table(index = ['userid'], columns = ['foodname'], values = 'rating')
Food_ratings.head()
```

```python
Food_ratings = Food_ratings.dropna(thresh = 3, axis = 1).fillna(0)
Food_ratings.head(20)
```

```python
Food_similarity_df = Food_ratings.corr(method = 'pearson')
Food_similarity_df.head()
```

```python
def Get_similar_foods(foodname, rating):
    if foodname in Food_similarity_df:
        Similar_score = Food_similarity_df[foodname]*(rating-2.5)
        Similar_score = Similar_score.sort_values(ascending = False)
        return Similar_score
    else:
        return pd.DataFrame()
```

```python
def Food_Recommender():
    user_id = int(input("Please enter your User ID: "))
    User_Foods = Food_table[Food_table["userid"] == user_id]
    Similar_Foods = pd.DataFrame()
    for i in User_Foods.itertuples():
        Similar_Foods = Similar_Foods.append(Get_similar_foods(i[-1], i[-2]), ignore_index = True)
    print("List of recommeneded foods you might like: ")
    print(Similar_Foods.sum().sort_values(ascending = False).head())
Food_Recommender()
```
